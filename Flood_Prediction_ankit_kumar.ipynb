# 2. Import Libraries


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

# 3. Load Dataset

```python
# Membaca dataset
data = pd.read_csv("flood.csv")
print(f'flood.csv : {data.shape}')
data.head(3)
```

# 4. Pre Processing

```python
X = data.drop(labels=['FloodProbability'], axis=1)
Y = data['FloodProbability']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20, random_state=42)
```

# 5. EDA
Mengidentifikasi jenis data, ukuran dataset, dan distribusi variabel untuk mengetahui karakteristiknya.

```python
print("Lima baris pertama dataset:")
print(data.head())
print("\nStatistik deskriptif:\n", data.describe())
```

KORELASI DENGAN HEATMAP

```python
corr = data.corr()
fig, ax = plt.subplots(figsize=(15, 10))
sns.heatmap(corr, annot=True, linewidths=.5, ax=ax)
plt.title("Heatmap Korelasi Fitur")
plt.show()
```

# 6.MODELLING DENGAN RANDOM FOREST DAN SVM

RANDOM FOREST

```python
# Model Random Forest
model_rf = RandomForestRegressor()
model_rf.fit(X_train, Y_train)
y_pred_rf = model_rf.predict(X_test)
```

```python
# Evaluasi Random Forest
print("\nEvaluasi Model Random Forest:")
print('Accuracy: ', model_rf.score(X_test, Y_test))
r2_rf = r2_score(Y_test, y_pred_rf)
mae_rf = mean_absolute_error(Y_test, y_pred_rf)
mse_rf = mean_squared_error(Y_test, y_pred_rf)
print("MSE:", mse_rf)
print("MAE:", mae_rf)
print("R2 SCORE:", r2_rf)
```

SVM

```python
model_svm = SVR(kernel='rbf')
model_svm.fit(X_train, Y_train)
y_pred_svm = model_svm.predict(X_test)
```

```python
# Evaluasi SVM
print("\nEvaluasi Model SVM:")
print('Accuracy: ', model_svm.score(X_test, Y_test))
r2_svm = r2_score(Y_test, y_pred_svm)
mae_svm = mean_absolute_error(Y_test, y_pred_svm)
mse_svm = mean_squared_error(Y_test, y_pred_svm)
print("MSE:", mse_svm)
print("MAE:", mae_svm)
print("R2 SCORE:", r2_svm)
```
